# Visual Studio 创建工程指南

在使用 Visual Studio 创建和管理工程时，有几个重要的注意事项，以确保项目结构清晰，易于维护和管理。

## 1. 创建工程

- 选择“空项目”选项，以便从头开始设置项目。
- 确保不使用预编译头文件（Precompiled Headers），以保持项目的简洁性和灵活性。

## 2. 目录结构

为了保持项目的清晰和可管理性，建议将目录分开保存：

- **编译目录**: 保存编译生成的中间文件和输出文件。示例：`project.win32`
- **源码目录**: 保存源代码文件。示例：`src`
- **运行时目录**: 保存运行时生成的文件，如可执行文件和日志文件。示例：`bin`

## 3. 配置调试器

- 设置调试器的工作目录为运行时目录，以确保调试器在运行程序时使用正确的工作目录。

## 4. 源码管理

- 手动在源码目录中创建源文件，并将其添加到 Visual Studio 项目中。
- 注意保持源码路径和文件夹分类的一致性。

## 5. 文件筛选器

- 在 Visual Studio 中使用筛选器来组织项目中的文件。
- 确保筛选器结构与源码目录结构保持一致，以便于项目管理。

## 示例目录结构

- **编译目录**: `project.win32`
- **源码目录**: `src`
- **运行时目录**: `bin`

这种结构有助于保持项目的整洁性，简化管理和维护工作。


# Visual Studio 编译注意事项

在使用 Visual Studio 编译代码时，有几个关键点需要注意，以确保编译过程顺利进行。

## 1. 使用快捷键进行编译

- **快捷键**: 使用 `Ctrl + Shift + B` 来启动代码编译。这将编译整个解决方案并生成可执行文件。
- **优势**: 快捷键可以提高编译效率，并确保所有文件都被编译，减少遗漏。

## 2. 编译单个文件

- **文件类型**: 当编译单个源文件（.c 或 .cpp 文件）时，生成的文件是 `.obj` 文件（对象文件）。
- **功能**: `.obj` 文件是程序编译后的二进制文件，包含中间代码。这些文件在链接阶段会与其他文件链接生成最终的可执行文件（.exe）。
- **别名**: `.obj` 文件也被称为目标文件或中间文件，提供了程序的相对地址信息。

## 3. 编译错误

- **常见问题**: 如果编译单个 `.c` 文件时出现错误，通常是由于代码不符合 C 语言的语法规范。
- **错误类型**: 包括语法错误、数据类型不匹配、未声明的变量等。
- **解决办法**: 解决编译错误需要检查代码的语法和结构，确保符合编程语言的规则。

## 4. 链接错误

- **问题描述**: 编译完单个源文件后，生成的 `.obj` 文件可能会在链接阶段产生错误。
- **常见原因**: 链接错误通常是由于找不到所需的依赖项，如库文件或其他目标文件。可能的原因包括缺少库文件、库文件路径不正确、符号未定义等。
- **解决办法**: 解决链接错误需要检查项目的依赖配置，确保所有必需的文件都被正确引用。



# 良好的编程习惯常用快捷方式

在编程过程中，掌握一些常用的快捷方式可以提高代码编写的效率和准确性。以下是一些常见的快捷键及其功能：

## 1. `Ctrl + C`
- **功能**: 复制选中的文本或文件到剪贴板。
- **用途**: 常用于复制代码片段或其他文本数据。

## 2. `Ctrl + V`
- **功能**: 粘贴剪贴板中的文本或文件到当前位置。
- **用途**: 将复制的内容插入到代码中。

## 3. `Ctrl + 左箭头 / 右箭头`
- **功能**: 将光标跳至当前单词的开头或结尾。
- **用途**: 左箭头跳到前一个单词的开头，右箭头跳到下一个单词的开头。方便在代码中快速移动到特定的单词。

## 4. `Ctrl + Shift + 左箭头 / 右箭头`
- **功能**: 选中从当前光标位置到前一个单词开头或下一个单词开头之间的文本。
- **用途**: 左箭头选中前一个单词，右箭头选中下一个单词的文本。用于快速选择多个单词。

## 5. `Shift + 上箭头 / 下箭头`
- **功能**: 逐行选择文本。
- **用途**: 按住 Shift 键并按上箭头或下箭头可以扩展或减少选中的行。用于逐行选择代码或文本块。

## 6. `Home`
- **功能**: 将光标移动到当前行的开头。
- **用途**: 便于快速返回行的起始位置。

## 7. `End`
- **功能**: 将光标移动到当前行的末尾。
- **用途**: 便于快速跳转到行的结束位置。

## 8. `Shift + Home`
- **功能**: 从当前光标位置选择到行的开头。
- **用途**: 用于快速选中从光标到行首的所有文本。

## 9. `Shift + End`
- **功能**: 从当前光标位置选择到行的末尾。
- **用途**: 用于快速选中从光标到行尾的所有文本。



# Windows 平台程序常见文件夹和文件

在 Windows 平台上，有许多常见的文件类型和文件夹，它们在程序开发和运行中扮演重要角色。以下是一些常见文件类型及其说明：

## 1. `.ini` 文件 — 初始化文件

- **用途**: `.ini` 文件用于存储系统和程序配置。它们通常包含程序的初始化参数和设置。
- **特性**:
  - 使用分号（`;`）作为注解标记。分号后的文字直到行尾都视为注解，不会被处理。
  - 示例内容：
    ```ini
    [SectionName]
    Key=Value
    ; This is a comment
    ```
    

## 2. `bin` 文件夹

- **用途**: `bin` 文件夹通常用于存放可执行的二进制文件，如程序的主执行文件。
- **说明**: `bin` 是 "binary" 的缩写，表示存放二进制文件的文件夹。

## 3. `.lib` 文件和 `.dll` 文件

### `.lib` 文件（静态链接库）

- **用途**: `.lib` 文件用于静态链接，在编译阶段将函数和变量的地址信息嵌入到最终的可执行文件中。
- **特性**:
  - 静态链接库的代码在编译时直接包含在生成的 EXE 文件中，不需要在运行时加载。
  - 示例：`example.lib`

### `.dll` 文件（动态链接库）

- **用途**: `.dll` 文件在程序运行时动态加载，提供函数、变量和类的实现。
- **特性**:
  - 动态链接库允许程序在运行时根据需要加载和卸载库文件，提高了模块化和可维护性。
  - 示例：`example.dll`

### 静态链接与动态链接的区别

1. **静态链接**:
   - 代码在编译阶段直接包含在最终生成的可执行文件中。
   - 优点：无需在运行时加载库，适用于简单的应用程序。
   - 缺点：增加了生成文件的大小，所有依赖都被包含在一个文件中。

2. **动态链接**:
   - 程序在运行时动态加载库文件，EXE 文件不包含所有库代码。
   - 优点：减少了可执行文件的大小，可以在运行时更新库而不重新编译应用程序。
   - 缺点：需要处理 DLL 文件的路径和版本兼容性问题。

### 链接过程

- **编译和链接**: 程序编译和运行涉及两个主要过程：
  1. **编译**: 将源代码转换为对象文件（.obj）。
  2. **链接**: 将对象文件与库文件中的符号进行关联，生成最终的可执行文件（.exe）。

- **链接类型**:
  - **静态链接**: 将所有需要的代码直接包含到可执行文件中。
  - **动态链接**: 仅在运行时加载所需的库文件（DLL）。

## 其他相关说明

- **动态/静态库的通用性**:
  - 动态库和静态库的概念是通用的，不仅限于 C/C++ 语言。
  - Windows 平台上，动态库使用 `.dll` 后缀，静态库使用 `.lib` 后缀。
  - 在 Linux 平台上，动态库使用 `.so` 后缀，静态库使用 `.a` 后缀。
  - Java 程序通过 JNI 可以调用本地库，但 Java 本身通过虚拟机运行，不直接涉及这些库。

这些文件和文件夹在 Windows 平台上是程序开发和运行的重要组成部分，了解它们的用途和特性可以帮助你更好地管理和维护你的项目。


# 程序启动过程

## 程序的启动过程

程序的启动过程可以分为以下几个步骤：

1. **操作系统创建一个进程**  
   操作系统为程序创建一个新的进程，并分配必要的资源。

2. **加载可执行文件**  
   操作系统将 `.exe` 文件的二进制代码加载到进程的代码段中。代码段是只读的，不能被修改。

3. **找到入口指令**  
   加载器根据程序的头部信息找到代码的入口指令，即程序的起始执行点。

4. **执行指令**  
   程序从入口指令开始，逐条执行程序中的指令。

5. **进程结束**  
   执行完所有指令后，进程终止，释放分配的资源。

   


## C语言入口函数定义

C 语言的入口函数是 `main` 函数，其定义如下：

argc: 操作系统传递给程序的参数个数。

argv: 操作系统传给我们的一个字符串数组，包含所有传递给程序的参数。argv 是 char* 类型的数组，其定义为 char**。

```c
#include <stdio.h>
#include <string.h>
#include <stdlib.h>

int main(int argc, char** argv) {
    printf("argc = %d\n", argc);
    for (int i = 0; i < argc; i++) {
        printf("argv[%d] = %s\n", i, argv[i]);
    }
    system("pause");
    return 0;
}
