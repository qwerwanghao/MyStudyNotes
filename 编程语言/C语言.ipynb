# 学习笔记：C 语言格式化输出示例

```c
#include <stdio.h>

int main() {
    // 整数输出
    int num = 42;
    printf("The number is: %d\n"); // 有符号十进制整数
    printf("The number (decimal): %i\n", num); // 另一种表示有符号十进制整数

    // 无符号整数输出
    unsigned int unum = 42;
    printf("The unsigned number is: %u\n", unum); // 无符号十进制整数

    // 八进制输出
    printf("The octal number is: %o\n", num); // 八进制整数

    // 十六进制输出
    printf("The hexadecimal number is: %x\n", num); // 十六进制整数（小写字母）
    printf("The hexadecimal number (uppercase) is: %X\n", num); // 十六进制整数（大写字母）

    // 32 位无符号整数输出
    unsigned long ul32 = 4294967295U; // 最大值为 2^32 - 1
    printf("The 32-bit unsigned integer is: %lu\n", ul32); // 32 位无符号整数

    // 64 位无符号整数输出
    unsigned long long ull64 = 18446744073709551615ULL; // 最大值为 2^64 - 1
    printf("The 64-bit unsigned integer is: %llu\n", ull64); // 64 位无符号整数

    // 浮点数输出
    float fnum = 3.14159;
    printf("The float number is: %f\n", fnum); // 浮点数

    // 科学计数法输出
    printf("The scientific notation is: %e\n", fnum); // 科学计数法（小写字母）
    printf("The scientific notation (uppercase) is: %E\n", fnum); // 科学计数法（大写字母）

    // %g 输出
    printf("The number in shortest representation: %g\n", fnum); // 根据实际值选择更简洁的输出格式（%e 或 %f）

    // 字符输出
    char ch = 'A';
    printf("The character is: %c\n", ch); // 单个字符

    // 字符串输出
    char str[] = "Hello, World!";
    printf("The string is: %s\n", str); // 字符串

    // 百分号输出
    printf("The percentage symbol: %%\n"); // 百分号

    // 宽度和精度控制
    printf("The number with width 10: %10d\n", num); // 输出的整数占据 10 个字符宽度，不足部分用空格填充
    printf("The float number with 2 decimal places: %.2f\n", fnum); // 浮点数小数点后保留 2 位

    // 左对齐
    printf("Left aligned number: %-10dend\n", num); // 左对齐的整数，宽度为 10

    // 指针地址输出
    printf("The address of num is: %p\n", (void*)&num); // 指针地址（需要强制转换为 void*）

    // 指定最小宽度和精度
    printf("Number with minimum width and precision: %6.3f\n", fnum); // 总宽度为 6，小数点后 3 位

    // 自定义格式化
    printf("Custom formatted float: %05.2f\n", fnum); // 总宽度为 5，浮点数保留 2 位小数，前导零

    return 0;
}


# 学习笔记：进制介绍与转换示例

```c
#include <stdio.h>

int main() {
    // 进制介绍

    // 二进制
    // 定义: 每位只能用0或1表示
    // 位值: 2^0, 2^1, 2^2, ...
    // 示例: 1101 (从右到左每位分别代表 2^0, 2^1, 2^2, 2^3)
    printf("Binary: Each bit is 0 or 1.\n");
    printf("Example: 1101 (from right to left: 2^0, 2^1, 2^2, 2^3)\n");

    // 十进制
    // 定义: 使用0-9这十个数字表示
    // 位值: 10^0, 10^1, 10^2, ...
    // 示例: 2024 = 4 * 10^0 + 2 * 10^1 + 0 * 10^2 + 2 * 10^3
    printf("Decimal: Uses digits 0-9.\n");
    printf("Example: 2024 = 4 * 10^0 + 2 * 10^1 + 0 * 10^2 + 2 * 10^3\n");

    // 十六进制
    // 定义: 每位可以用[0, 15]来表示
    // 表示方法:
    // 0-9: 保持不变
    // 10-15: 用A-F表示
    // 示例: 0, 1, 2, ..., 9, A, B, C, D, E, F
    printf("Hexadecimal: Each digit represents [0, 15].\n");
    printf("Digits: 0, 1, 2, ..., 9, A, B, C, D, E, F\n");

    // 进制转换

    // 二进制到十进制的转换
    // 方法: 将每位的值乘以其对应的2的幂，然后求和
    // 示例: 1101(二进制) = 1 * 2^3 + 1 * 2^2 + 0 * 2^1 + 1 * 2^0 = 8 + 4 + 0 + 1 = 13(十进制)
    printf("Binary to Decimal: Multiply each bit by corresponding power of 2 and sum.\n");
    printf("Example: 1101 (binary) = 1 * 2^3 + 1 * 2^2 + 0 * 2^1 + 1 * 2^0 = 13 (decimal)\n");

    // 十进制到二进制
    // 方法: 反复除以2，取余数，直到商为0，然后从下往上读取余数
    // 示例: 13 ÷ 2 = 6 余 1 6 ÷ 2 = 3 余 0 3 ÷ 2 = 1 余 1 1 ÷ 2 = 0 余 1
    // 因此，13(十进制) = 1101(二进制)
    printf("Decimal to Binary: Repeatedly divide by 2, record remainders, read remainders bottom to top.\n");
    printf("Example: 13 (decimal) = 1101 (binary)\n");

    // 二进制到十六进制
    // 方法: 将二进制数每4位分组，把每组转换为对应的十六进制数字
    // 示例: 1101 0011(二进制) → 1101 0011 → D3(十六进制)
    printf("Binary to Hexadecimal: Group binary digits in 4s, convert each group to hex.\n");
    printf("Example: 1101 0011 (binary) = D3 (hexadecimal)\n");

    // 十六进制到二进制
    // 方法: 将每个十六进制数字转换为4位二进制
    // 示例: D3(十六进制) → 1101 0011(二进制)
    printf("Hexadecimal to Binary: Convert each hex digit to 4-bit binary.\n");
    printf("Example: D3 (hexadecimal) = 1101 0011 (binary)\n");

    // 十进制到十六进制
    // 方法: 反复除以16，取余数，直到商为0，然后从下往上读取余数
    // 示例: 2748 ÷ 16 = 171 余 12 (C) 171 ÷ 16 = 10 余 11 (B) 10 ÷ 16 = 0 余 10 (A)
    // 因此，2748(十进制) = ABC(十六进制)
    printf("Decimal to Hexadecimal: Repeatedly divide by 16, record remainders, read remainders bottom to top.\n");
    printf("Example: 2748 (decimal) = ABC (hexadecimal)\n");

    // 十六进制到十进制
    // 方法: 将每位的值乘以16的幂次方，然后求和
    // 示例: ABC(十六进制) = 10 * 16^2 + 11 * 16^1 + 12 * 16^0 = 2560 + 176 + 12 = 2748(十进制)
    printf("Hexadecimal to Decimal: Multiply each digit by power of 16 and sum.\n");
    printf("Example: ABC (hexadecimal) = 2748 (decimal)\n");

    return 0;
}


# 学习笔记：数据单位介绍

```c
#include <stdio.h>

int main() {
    // 位 (Bit)
    // 位是计算机数据存储的最小单位。每位可以存储 0 或 1。
    printf("Bit: 0 or 1\n");

    // 字节 (Byte)
    // 一个字节由 8 个位组成。字节是数据存储的基本单位。
    // 1 字节 = 8 位 = 256 种不同的值（0 到 255）
    printf("Byte: 8 bits, %d possible values (0 to 255)\n", 256);

    // 字 (Word)
    // 一个字通常由 2 个字节组成，即 16 位。
    // 1 字 = 2 字节 = 16 位 = 65,536 种不同的值（0 到 65,535）
    printf("Word: 16 bits, %d possible values (0 to 65,535)\n", 65536);

    // 双字 (Double Word)
    // 一个双字由 4 个字节组成，即 32 位。
    // 1 双字 = 4 字节 = 32 位 = 4,294,967,296 种不同的值（0 到 4,294,967,295）
    printf("Double Word: 32 bits, %llu possible values (0 to %llu)\n", 4294967296ULL, 4294967295ULL);

    // 四字 (Quad Word)
    // 一个四字由 8 个字节组成，即 64 位。
    // 1 四字 = 8 字节 = 64 位 = 18,446,744,073,709,551,616 种不同的值（0 到 18,446,744,073,709,551,615）
    printf("Quad Word: 64 bits, %llu possible values (0 to %llu)\n", 18446744073709551616ULL, 18446744073709551615ULL);

    return 0;
}


# 学习笔记：计算机的数据表示

```c
#include <stdio.h>
#include <stdint.h>

// 学习笔记：计算机的数据表示

int main() {
    // 无符号整数
    // 定义: 所有的位（bit）都用来表示数据的大小，不带符号，因此只能表示非负数。
    // 示例: 8位无符号整数的范围是 0 到 255 (2^8 - 1)。
    uint8_t unsignedInt = 200; // 8位无符号整数
    printf("Unsigned Integer Example:\n");
    printf("Value: %u\n", unsignedInt);

    // 有符号数
    // 定义: 最高位（sign bit）表示符号，剩余的位表示数据值。
    // - 正数: 最高位是0，数据部分是数值的二进制表示。
    // - 负数: 最高位是1，数据部分是正数的二进制表示取反再加1（补码表示）。
    // 示例: 对于8位整数，正数3的二进制是 0000 0011，负数-3的二进制是 1111 1101（取反再加1）。
    int8_t positiveInt = 3;
    int8_t negativeInt = -3;
    printf("\nSigned Integer Example:\n");
    printf("Positive Value: %d\n", positiveInt);
    printf("Negative Value: %d\n", negativeInt);
    printf("Sum of 3 + (-3) = %d\n", positiveInt + negativeInt);

    // 小数
    // 定义: 使用科学计数法表示小数，浮点数有单精度（4字节）和双精度（8字节）两种。
    // - 单精度浮点数: 32位，其中1位符号位，8位指数位，23位尾数位。
    // - 双精度浮点数: 64位，其中1位符号位，11位指数位，52位尾数位。
    float singlePrecision = 3.14f;  // 单精度浮点数
    double doublePrecision = 3.141592653589793; // 双精度浮点数
    printf("\nFloating Point Example:\n");
    printf("Single Precision: %.2f\n", singlePrecision);
    printf("Double Precision: %.15f\n", doublePrecision);

    // 二进制的解释
    // 二进制数据在计算机中可以被解释为不同的数据类型:
    // - 作为正数: 直接解释为正数。
    // - 作为负数: 使用补码表示负数。
    // - 作为无符号数据: 只看数据大小，不考虑符号。
    // - 作为小数: 按照浮点数表示。
    // 示例:
    //   0b1100 (二进制) 在不同解释下：
    //   - 正数: 12 (十进制)
    //   - 负数: -4 (在4位补码表示中)
    //   - 无符号数据: 12 (十进制)
    //   - 小数: 需要考虑浮点数格式
    int8_t binaryAsInt = 0b1100;
    printf("\nBinary Interpretation Example:\n");
    printf("Binary 0b1100 interpreted as integer: %d\n", binaryAsInt);

    return 0;
}


# 学习笔记：字符

```c
#include <stdio.h>

// 学习笔记：字符

int main() {
    // 计算机存储字符的基本概念
    // 1. 计算机内部没有直接存储字符的能力，只能存储0和1（二进制数据）。
    // 2. 因此，我们需要定义一种方法，将数据映射到对应的字符，这个过程称为编码。

    // 数据到字符的映射（编码）
    // 3. 编码是将二进制数据转换成可识别的字符的规则。常见的编码方式包括：
    
    // (1) ASCII编码
    // - 只用一个字节（8位）表示一个字符。
    // - 主要用于表示英文字符、数字和一些控制符。
    // - 示例: 'A' 的 ASCII 编码是 65，即 0x41。
    char asciiChar = 'A';
    printf("ASCII Encoding Example:\n");
    printf("Character 'A' in ASCII: %c (Decimal: %d, Hexadecimal: %X)\n", asciiChar, asciiChar, asciiChar);

    // (2) UTF-8编码
    // - 用于表示多语言字符，包括大部分文字和符号。
    // - 采用变长编码：字符可以用1到4个字节表示。
    // - 示例: '€' (Euro符号) 的 UTF-8 编码是 0xE2 0x82 0xAC。
    char utf8Char = '€'; // 注意：这在实际代码中可能需要额外处理以正确显示非ASCII字符
    printf("\nUTF-8 Encoding Example:\n");
    printf("Character '€' in UTF-8: %c\n", utf8Char);

    // (3) 其他编码
    // - 还有许多其他编码方式，例如 UTF-16、ISO-8859-1 等，根据不同的应用需求选择使用。
    // - 示例: UTF-16 是用16位编码表示字符，适合需要表示更多字符集的应用。

    // 字符的显示
    // 4. 尽管计算机内部不能直接表示字符，但我们可以通过编码将数据解释为字符。
    //    - printf 函数等标准库函数根据编码方式将数据转换为字符进行显示。
    //    - 计算机图形学也可以通过绘制算法将字符渲染到屏幕上。

    printf("\nCharacter Display Example:\n");
    printf("Displaying characters: %c %c\n", 'A', '€');

    return 0;
}


# 学习笔记：补码

```c
#include <stdio.h>
#include <stdint.h>

// 学习笔记：补码

int main() {
    // 补码概念
    // 1. 补码是一种用于表示有符号整数的二进制编码方式。
    // 2. 在补码表示中，最高位（即最左边的一位）表示符号位：
    //    - 符号位为0表示正数。
    //    - 符号位为1表示负数。

    // 3. 正数的补码与其二进制表示相同。
    // 4. 负数的补码计算方法：
    //    - 先将正数的二进制表示取反（即将所有0变为1，所有1变为0）。
    //    - 然后在取反后的结果上加1。

    // 示例：8位补码表示
    // 5. 正数示例
    int8_t positiveNum = 5; // 正数5的补码是其二进制表示 0000 0101
    printf("Positive Number Example:\n");
    printf("Decimal: %d, Binary: ", positiveNum);
    for (int i = 7; i >= 0; i--) {
        printf("%d", (positiveNum >> i) & 1);
    }
    printf("\n");

    // 6. 负数示例
    // 负数 -5 的补码计算过程
    // 1. 先计算 +5 的二进制：0000 0101
    // 2. 取反得到：1111 1010
    // 3. 加1得到：1111 1011
    int8_t negativeNum = -5; // 负数-5的补码是 1111 1011
    printf("\nNegative Number Example:\n");
    printf("Decimal: %d, Binary: ", negativeNum);
    for (int i = 7; i >= 0; i--) {
        printf("%d", (negativeNum >> i) & 1);
    }
    printf("\n");

    // 7. 补码运算
    // 补码运算简化了加法和减法的处理，因为加法和减法可以用相同的硬件来实现。
    // 例如，计算 5 + (-5) 的补码运算：
    // 5 的补码是 0000 0101
    // -5 的补码是 1111 1011
    // 结果是：0000 0101 + 1111 1011 = 0000 0000 (进位被丢弃)
    printf("\nComplement Operation Example:\n");
    int8_t sum = positiveNum + negativeNum; // 应该是 0
    printf("Sum of %d and %d = %d\n", positiveNum, negativeNum, sum);

    return 0;
}


# 学习笔记：计算机中的大小端、内存地址、内存访问顺序

```c
#include <stdio.h>

// 学习笔记：计算机中的大小端、内存地址、内存访问顺序

int main() {
    // 大小端（Endianness）
    // 1. 大小端表示数据在内存中的存储顺序。
    //    - 大端（Big Endian）：高位字节存储在低地址，低位字节存储在高地址。
    //    - 小端（Little Endian）：低位字节存储在低地址，高位字节存储在高地址。

    // 2. 示例：32位整数0x12345678的存储方式
    //    - 大端模式：0x12 0x34 0x56 0x78
    //    - 小端模式：0x78 0x56 0x34 0x12

    // 3. 确定大小端模式的规则取决于硬件平台和操作系统架构。
    //    - 硬件平台的架构、操作系统、应用程序需求、数据传输需求等因素会影响选择。

    // 大小端转换代码示例（C#）
    // 1. 将32位整数从小端序转换为大端序
    // 2. 代码示例：
    //    int value = 0x12345678; // 小端序表示
    //    byte[] bytes = BitConverter.GetBytes(value);
    //    Array.Reverse(bytes); // 反转字节数组
    //    int result = BitConverter.ToInt32(bytes, 0); // 大端序表示
    // 3. 注意：大小端转换适用于多字节数据，单字节数据（如byte、char）不需要转换。

    // 大端和小端优缺点
    // 大端优点：
    //    1. 不需转换字节序，适合不同字节序数据通信。
    //    2. 按自然排序，易于理解。
    // 大端缺点：
    //    1. 处理时需位移运算，效率较低。
    //    2. 网络传输需转换字节序，增加开销。
    // 小端优点：
    //    1. 处理器内部直接使用低位字节，提高效率。
    //    2. 减少字节序转换次数，通常更高效。
    // 小端缺点：
    //    1. 需转换字节序，不适合不同字节序数据通信。
    //    2. 不符合自然排序，增加学习成本。

    // 游戏开发中的字节序选择
    // 1. 网络协议：遵循网络协议规定的字节序。
    // 2. 平台兼容性：确保游戏在所有平台上正常工作。
    // 3. 数据格式：根据文件格式和数据结构选择字节序。
    // 4. 性能：考虑字节序对内存访问和数据传输的影响。

    return 0;
}


# 学习笔记：位运算

```c
#include <stdio.h>

// 学习笔记：位运算

int main() {
    // 位运算操作简介
    // 位运算可以对二进制数的每一位进行操作，包括按位与、按位或、按位异或、按位取反、左移和右移。
    // 这些操作常用于代码优化、加密算法、图像处理等领域。

    // a. 按位与（&）
    // 对两个二进制数的每一位进行与运算，只有当两个数的对应位都为1时，结果的对应位才为1，否则为0。
    // 示例：1010 & 1100 = 1000
    unsigned char a = 0b1010;
    unsigned char b = 0b1100;
    unsigned char result_and = a & b;
    printf("按位与操作:\n");
    printf("0b1010 & 0b1100 = 0b%04b\n", result_and);

    // b. 按位或（|）
    // 对两个二进制数的每一位进行或运算，只有当两个数的对应位都为0时，结果的对应位才为0，否则为1。
    // 示例：1010 | 1100 = 1110
    unsigned char result_or = a | b;
    printf("\n按位或操作:\n");
    printf("0b1010 | 0b1100 = 0b%04b\n", result_or);

    // c. 按位异或（^）
    // 对两个二进制数的每一位进行异或运算，只有当两个数的对应位不同时，结果的对应位才为1，否则为0。
    // 示例：1010 ^ 1100 = 0110
    unsigned char result_xor = a ^ b;
    printf("\n按位异或操作:\n");
    printf("0b1010 ^ 0b1100 = 0b%04b\n", result_xor);

    // d. 按位取反（~）
    // 对一个二进制数的每一位进行取反运算，即0变成1，1变成0。
    // 示例：~1010 = 0101（假设8位数）
    unsigned char result_not = ~a;
    printf("\n按位取反操作:\n");
    printf("~0b1010 = 0b%04b\n", result_not & 0x0F); // 只显示低4位

    // e. 左移（<<）
    // 将一个二进制数的所有位向左移动指定的位数，右侧空出的位用0填充。
    // 示例：1010 << 2 = 101000
    unsigned char result_left_shift = a << 2;
    printf("\n左移操作:\n");
    printf("0b1010 << 2 = 0b%06b\n", result_left_shift); // 结果需要6位显示

    // f. 右移（>>）
    // 将一个二进制数的所有位向右移动指定的位数，左侧空出的位用符号位填充（对于有符号数）或0填充（对于无符号数）。
    // 示例：1010 >> 2 = 0010
    unsigned char result_right_shift = a >> 2;
    printf("\n右移操作:\n");
    printf("0b1010 >> 2 = 0b%04b\n", result_right_shift);

    // g. 注意事项
    // - 左移和右移的位数不能超过数据类型的位数，否则会导致数据溢出或未定义行为。
    // - 例如，对于8位二进制数，左移9位或右移9位的结果是未定义的。
    // - 在进行位移运算时，要注意数据类型的位数和移动的位数，以避免数据溢出和错误结果。

    return 0;
}
