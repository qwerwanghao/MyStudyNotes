C# 中的 **Collections**（集合）类提供了许多处理数据的工具，用于不同的场景。它们并不都基于数组实现，而是根据用途选择了不同的数据结构（如链表、哈希表、树等）作为底层实现。

---

### **常用的 C# Collections 类**

C# 中的集合分为以下几类，每类都有自己的特点和底层实现：

---

#### **1. 基础集合类**

- **`ArrayList`** (非泛型)
  - **描述**：动态数组，容量可以自动调整。
  - **底层实现**：基于数组。
  - **缺点**：非泛型，类型安全差。
  - **替代方案**：`List<T>`。

- **`List<T>`** (泛型)
  - **描述**：动态大小的类型安全数组。
  - **底层实现**：基于数组，容量不足时会自动扩容（复制到一个更大的数组）。
  - **用途**：需要快速随机访问和动态大小数组。

---

#### **2. 映射（键值对）集合**

- **`Dictionary<TKey, TValue>`**
  - **描述**：键值对集合，允许快速查找值。
  - **底层实现**：基于哈希表（`Hashtable` 的泛型实现）。
  - **用途**：通过键快速查找值。

- **`SortedDictionary<TKey, TValue>`**
  - **描述**：根据键自动排序的字典。
  - **底层实现**：基于红黑树。
  - **用途**：需要保持键排序。

- **`SortedList<TKey, TValue>`**
  - **描述**：键值对集合，按键排序，使用内存更高效。
  - **底层实现**：基于两个并行数组（一个存储键，一个存储值）。
  - **用途**：需要有序存储且查找频率较高。

- **`Hashtable`** (非泛型)
  - **描述**：非泛型键值对集合。
  - **底层实现**：基于哈希表。
  - **缺点**：非类型安全，已被 `Dictionary<TKey, TValue>` 替代。

---

#### **3. 集合类**

- **`HashSet<T>`**
  - **描述**：无序、不重复的集合。
  - **底层实现**：基于哈希表。
  - **用途**：快速检查元素是否存在。

- **`SortedSet<T>`**
  - **描述**：自动排序且不重复的集合。
  - **底层实现**：基于红黑树。
  - **用途**：需要有序存储且不重复。

---

#### **4. 队列与栈**

- **`Queue<T>`**
  - **描述**：先进先出（FIFO）集合。
  - **底层实现**：基于循环数组。
  - **用途**：队列操作，如任务调度。

- **`Stack<T>`**
  - **描述**：后进先出（LIFO）集合。
  - **底层实现**：基于动态数组。
  - **用途**：栈操作，如递归调用或回溯算法。

---

#### **5. 链表类**

- **`LinkedList<T>`**
  - **描述**：双向链表。
  - **底层实现**：基于链表节点。
  - **用途**：需要频繁插入和删除操作。

---

#### **6. 并发集合类**

- **`ConcurrentDictionary<TKey, TValue>`**
  - **描述**：线程安全的字典。
  - **底层实现**：基于分段锁的哈希表。
  - **用途**：多线程环境下的键值对存储。

- **`ConcurrentQueue<T>`**
  - **描述**：线程安全的队列。
  - **底层实现**：基于链表。
  - **用途**：线程安全的任务队列。

- **`ConcurrentStack<T>`**
  - **描述**：线程安全的栈。
  - **底层实现**：基于链表。
  - **用途**：线程安全的后进先出操作。

- **`ConcurrentBag<T>`**
  - **描述**：线程安全的无序集合。
  - **底层实现**：每个线程拥有自己的局部存储。
  - **用途**：线程安全的非有序数据存储。

---

#### **7. 不可变集合类**

- **`ImmutableList<T>`**
  - **描述**：一旦创建后无法修改的列表。
  - **底层实现**：基于持久化数据结构。
  - **用途**：需要数据不可变的场景（如并发）。

- **`ImmutableDictionary<TKey, TValue>`**
  - **描述**：一旦创建后无法修改的字典。
  - **底层实现**：基于持久化哈希表。

- **`ImmutableHashSet<T>`**
  - **描述**：一旦创建后无法修改的集合。
  - **底层实现**：基于持久化哈希表。

---

### **是否基于数组实现**

并非所有集合类都基于数组实现，不同集合类使用不同的数据结构：

| 集合类型                   | 底层实现                     |
|---------------------------|-----------------------------|
| **数组（Array）**           | 数组                       |
| **List<T>**               | 动态数组                   |
| **Dictionary<TKey, TValue>** | 哈希表                     |
| **Queue<T>**              | 循环数组                   |
| **Stack<T>**              | 动态数组                   |
| **HashSet<T>**            | 哈希表                     |
| **SortedSet<T>**          | 红黑树                     |
| **LinkedList<T>**         | 链表节点                   |
| **ConcurrentQueue<T>**    | 链表                       |
| **ImmutableList<T>**      | 持久化数据结构（非数组）       |

---

### **如何选择合适的集合**

1. **需要快速访问**：
   - **选择**：`Array` 或 `List<T>`。
   - **原因**：基于数组，支持 O(1) 的索引访问。

2. **需要快速查找或检查存在性**：
   - **选择**：`Dictionary<TKey, TValue>` 或 `HashSet<T>`。
   - **原因**：基于哈希表，支持 O(1) 的查找。

3. **需要排序**：
   - **选择**：`SortedDictionary<TKey, TValue>` 或 `SortedSet<T>`。
   - **原因**：基于红黑树，支持有序存储。

4. **需要频繁插入和删除**：
   - **选择**：`LinkedList<T>`。
   - **原因**：基于链表，无需移动元素。

5. **多线程场景**：
   - **选择**：`ConcurrentDictionary<TKey, TValue>` 或 `ConcurrentQueue<T>`。
   - **原因**：线程安全。

6. **数据不可变**：
   - **选择**：`ImmutableList<T>`、`ImmutableDictionary<TKey, TValue>`。
   - **原因**：防止数据在创建后被修改。

---

### **总结**

C# 提供了丰富的集合类型，基于不同的底层实现（如数组、哈希表、链表、红黑树等）。选择集合类型时，应根据需求（如访问速度、内存占用、是否需要线程安全）进行权衡和选择。
